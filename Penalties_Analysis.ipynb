```
#count 
df['penaltyCodes'].value_counts()
```

```
#keeping relevant 

week1 = df[['penaltyCodes','playResult','coverage','quarter','down','yardsToGo','defendersInTheBox', 'numberOfPassRushers', 'personnelD','offenseFormation', 'personnelO']]

week1.value_counts()
```

```
#Drop Na Rows 

week1.dropna(subset = ['penaltyCodes','coverage','playResult','quarter','down','yardsToGo','defendersInTheBox', 'numberOfPassRushers', 'personnelD','offenseFormation', 'personnelO'], inplace=True)

week1.reset_index(drop=True)
```

```
print(week1.columns)
```


# converting using ML 

```
week1.columns
```

```
#label Encoding
import sklearn
from sklearn.preprocessing import LabelEncoder

#Making Label 
lb_make = LabelEncoder()

#Penalty codes Encoding 
week1['PT'] = lb_make.fit_transform(week1['penaltyCodes'])
# Converting number of pass rushers 
week1['PR'] = lb_make.fit_transform(week1['numberOfPassRushers'])
#PersonelD 
week1['DP'] = lb_make.fit_transform(week1['personnelD'])
#offenseFormation 
week1['OF'] = lb_make.fit_transform(week1['offenseFormation'])
#offense Personnell
week1['PO'] = lb_make.fit_transform(week1['personnelO'])
#Coverage
week1['C'] = lb_make.fit_transform(week1['coverage'])
#Results 
week1.head()
```

```
W1 = week1[['playResult', 'quarter', 'down','yardsToGo', 'defendersInTheBox', 'numberOfPassRushers','PT', 'PR', 'DP', 'OF', 'PO', 'C']]

W1.reset_index
W1.head()
```
